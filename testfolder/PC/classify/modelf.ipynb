```
import numpy as np
import math
from sklearn.model_selection import train_test_split
import keras.backend as K
from tensorflow.keras.layers import Conv1D, MaxPool1D, Dropout, Flatten, Dense
from tensorflow.keras.models import *
from tensorflow.keras import optimizers
from sklearn.preprocessing import StandardScaler

def dh12(x):
	temp=np.zeros(x.shape)
	temp[:]=x[:]
	for i in range(len(x[:,-1])-2):
		x[i+1,-1]=int((temp[i+2,-1]-temp[i,-1])/2)
	return np.expand_dims(x[1:len(x)-1,-1], 1)
def standardd(X):
	Xnew=np.asarray(X,dtype='float')
	total=0
	u=X.sum()/len(X)
	for i in range(len(X)):
		total+=((X[i]-u)**2)
	o=math.sqrt(total/len(X))
	print(o)
	for i in range(len(X)):
		Xnew[i]=float((X[i]-u)/o)
	return Xnew
def processdata():
  training = np.load('drive/My Drive/ML_tutorial/training.npz')
  X_data = training['xdata']
  y_true = training['ytrue']
  X_datanew=[]
  for i in range(X_data.shape[0]):
	  X=dh12(X_data[i])
	  X_datanew.append(X)
  X_datanew=np.asarray(X_datanew)
  for i in range(X_datanew.shape[0]):
	  X_datanew[i]=standardd(X_datanew[i])
  y_onehot = np.zeros((len(y_true), 9))
  y_onehot[np.arange(len(y_true)), y_true] = 1
  return X_datanew,y_onehot

model = Sequential()
model.add(Conv1D(16, 2, activation = 'relu', input_shape = (226, 1)))
model.add(MaxPool1D(pool_size=2))
#model.add(Dropout(0.2))
model.add(Conv1D(32, 2, activation = 'relu'))
model.add(MaxPool1D(pool_size=2))
#model.add(Dropout(0.2))
model.add(Conv1D(32, 2, activation = 'relu'))
model.add(MaxPool1D(pool_size=2))
#model.add(Dropout(0.2))
model.add(Conv1D(64, 2, activation = 'relu'))
model.add(MaxPool1D(pool_size=2))
#model.add(Dropout(0.2))
model.add(Flatten())
model.add(Dense(128, activation='relu'))
model.add(Dense(9, activation='softmax'))
adam = optimizers.Adam(0.0001, 0.9, 0.99)
model.compile(adam, 'categorical_crossentropy', ['accuracy'])

X_datanew,y_onehot=processdata()

X_train, X_test, y_train, y_test = train_test_split(X_datanew, y_onehot)
history = model.fit(X_train, y_train, 64, 2000, 1, validation_data=(X_test, y_test))
with open("drive/My Drive/ML_tutorial/Model/modeledit.json", 'w') as model_file:
  model_file.write(model.to_json())
model.save("drive/My Drive/ML_tutorial/Weight/modeledit.h5")
```
